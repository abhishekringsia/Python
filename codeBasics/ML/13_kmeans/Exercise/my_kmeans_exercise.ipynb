```python
import pandas as pd
import numpy as np 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data_set = load_iris()
dir(data_set)
```

```python
df = pd.DataFrame(data_set.data, columns= data_set.feature_names)
df.head()
```

```python
df = df.drop(["sepal length (cm)","sepal width (cm)"] ,axis = "columns")
```

```python
df.head()
```

```python
df["target"] = data_set.target
```

```python
df.head()
```

```python
%matplotlib inline
plt.scatter(df["petal length (cm)"], df["petal width (cm)"])
```

```python
model = KMeans(n_clusters=2)
```

```python
y_predicted = model.fit_predict(df[["petal length (cm)"]], df[["petal width (cm)"]])
```

```python
df["result"] = y_predicted
df.head()
```

```python
k_rng = range(1,10)
sse= []
for k in k_rng:
    md = KMeans(n_clusters=k)
    y_predicted =   md.fit_predict(df[["petal length (cm)"]], df[["petal width (cm)"]])
    sse.append(md.inertia_)
    
```

```python
sse
```

```python
plt.scatter(k_rng,sse)
```

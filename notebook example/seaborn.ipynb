```python
import seaborn as sns
import pandas as pd

wine_counts = pd.read_csv("winemag-data_first150k.csv", index_col=0)
wine_counts.head()
```

```python
sns.countplot(wine_counts["points"])
```

```python
sns.kdeplot(wine_counts.query("price < 200").price)
```

```python
sns.kdeplot(wine_counts["points"])
```

```python
#kernel density estimate, address fundamental weakness of line chart
sns.kdeplot(wine_counts.loc[:, ['price', 'points']].dropna().sample(5000))
```

```python
#distplot - pandas histogram is the distplot
sns.distplot(wine_counts['points'], bins=10 ,kde = False)
```

```python
sns.jointplot(x = "price" , y = "points" , data = wine_counts[wine_counts["price"] < 200] 
              )
```

```python
sns.jointplot(x = "price" , y = "points" , data = wine_counts[wine_counts["price"] < 200] 
              , kind ="hex",gridsize = 20)
```

```python
sns.boxplot(
    x='variety',
    y='points',
    data=wine_counts[wine_counts.variety.isin(wine_counts.variety.value_counts()[:5].index)]
)
```

```python
sns.violinplot(
    x='variety',
    y='points',
    data=wine_counts[wine_counts.variety.isin(wine_counts.variety.value_counts()[:5].index)]
)
```

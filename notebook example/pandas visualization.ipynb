```python
import matplotlib.pyplot as plt
import pandas as pd

reviews = pd.read_csv("winemag-data_first150k.csv", index_col=0)

reviews["province"].value_counts().head(5).plot.bar()
```

```python
(reviews["province"].value_counts().head(10)/len(reviews)).plot.bar()
```

```python
#histogram - split data in interwal and shows chart
reviews["price"].plot.hist()
```

```python
reviews[reviews["price"] < 100].sample(100).plot.scatter(x = "price" , y ="points")
```

```python
#hex plot aggregates points in space into hexagons, and then colors those hexagons based on the values within them
reviews[reviews["price"] < 100].plot.hexbin(x="price", y="points" ,gridsize =5)
```

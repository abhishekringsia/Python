```python
import pandas as pd 
india_weather = {"city" : ["mumbai", "delhi", "banglore","pune"],
                "tempreature" : [32,45,30,40],
                "humidity": [80,60,78,80],
                "city_type" : ["urban","urban","village","village"]}
india_weather = pd.DataFrame(india_weather)
india_weather
```

```python
pd.crosstab(india_weather.city,india_weather.city_type, margins = True)
```

```python
import numpy as np
pd.crosstab([india_weather.city],[india_weather.city_type],values = india_weather.humidity, aggfunc=np.average)
```
